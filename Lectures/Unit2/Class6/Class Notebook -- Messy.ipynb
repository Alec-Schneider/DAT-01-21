```python
sales['NumOne'] = 1
```

```python
sales.head()
```

```python
sales.drop(['NumOne'], axis=1, inplace=True)
```

```python
sales['Profit Margin'] = sales['Profit']/sales['Sales']
```

```python
sales['Profit Margin']
```

```python
sales['Profitable'] = np.where(sales['Profit'] > 0, 'Profitable', 'Not Profitable')
```

```python
sales['Profitable'].value_counts()
```

```python
sales['Profitable'].dtype
```

```python
conditions = [
    (sales['Region'] == 'East') & (sales['Profitable'] == 'Profitable'),
    (sales['Region'] == 'East') & (sales['Profitable'] == 'Not Profitable'),
    (sales['Region'] == 'West') & (sales['Profitable'] == 'Profitable')
]

results = ['East-Profitable', 'East-Not Profitable', 'West-Profitable']

sales['RegionProfits'] = np.select(conditions, results, 'Other')
```

```python
%timeit sales['RegionProfits'] = np.select(conditions, results, 'Other')
```

```python
sales['RegionProfits'].value_counts()
```

```python
def find_category(row):
    if row['Region'] == 'East':
        if row['Profitable'] == 'Profitable':
            return 'Yes!'
        else:
            return 'No!'
    elif row['Region'] == 'West':
        if row['Profitable'] == 'Profitable':
            return 'West is the best!'
        else:
            return 'West is not the best!'
    else:
        return 'Other'
```

```python
%timeit sales['OtherCol'] = sales.apply(find_category, axis=1)
```

```python
sales['Profitable'] = np.where(sales['Profit'] > 0, True, False)
```

```python
sales['Ship Mode']
```

```python
conditions = [
    sales['Ship Mode'] == 'Same Day',
    sales['Ship Mode'] == 'First Class',
    sales['Ship Mode'] == 'Second Class',
    sales['Ship Mode'] == 'Standard Class',
]

results = [0, 2, 3, 6]

sales['Expected Ship Time'] = np.select(conditions, results, -1)
```

```python
sales['Actual Ship Time'] = (sales['Ship Date'] - sales['Order Date']).dt.days
```

```python
sales['Late'] = np.where(sales['Actual Ship Time'] > sales['Expected Ship Time'], True, False)
```
